
<p>
ในกล่องใบหนึ่งบรรจุ ลูกบอลสีแดง 3 ลูก ลูกบอลสีน้ำเงิน 2 ลูก และลูกบอลสีเขียว จำนวน 5 ลูก โดยทำการสุ่มลูกบอลทีละหนึ่งลูกแล้วใส่คืนเพื่อสุ่มหยิบครั้งใหม่ จำนวน 10 ครั้ง สนใจความน่าจะเป็น ที่จะได้ลูกบอลสีน้ำเงินจำนวน 6 (ใน 10) ครั้ง
<p>

```python
import random

# กำหนดจำนวนลูกบอลแต่ละสี
red_balls = 3
blue_balls = 2
green_balls = 5

# จำนวนครั้งที่จะทำการสุ่ม
num_trials = 10

# จำนวนครั้งที่ได้ลูกบอลสีน้ำเงิน
blue_count = 0

# ฟังก์ชันสำหรับการสุ่มลูกบอล
def draw_ball():
    # สุ่มตัวเลขระหว่าง 1 ถึง 10
    rand_num = random.randint(1, 10)
    # ตรวจสอบสีของลูกบอล
    if rand_num <= red_balls:
        return "red"
    elif rand_num <= red_balls + blue_balls:
        return "blue"
    else:
        return "green"

# ทำการสุ่มและนับจำนวนลูกบอลสีน้ำเงิน
for _ in range(num_trials):
    ball_color = draw_ball()
    if ball_color == "blue":
        blue_count += 1

# คำนวณความน่าจะเป็นที่จะได้ลูกบอลสีน้ำเงิน 6 ครั้งใน 10 ครั้ง
probability = blue_count / num_trials

# แสดงผลลัพธ์
print("ผลลัพธ์การทดลองการสุ่มลูกบอล:")
print("================================")
print(f"จำนวนลูกบอลสีแดง: {red_balls} ลูก")
print(f"จำนวนลูกบอลสีน้ำเงิน: {blue_balls} ลูก")
print(f"จำนวนลูกบอลสีเขียว: {green_balls} ลูก")
print("--------------------------------")
print(f"จำนวนครั้งที่ทำการสุ่ม: {num_trials} ครั้ง")
print(f"จำนวนครั้งที่ได้ลูกบอลสีน้ำเงิน: {blue_count} ครั้ง")
print("--------------------------------")
print(f"ความน่าจะเป็นที่จะได้ลูกบอลสีน้ำเงิน 6 ครั้งใน 10 ครั้ง: {probability:.2f}")
```

```python
import random

# กำหนดจำนวนลูกบอลแต่ละสี
red_balls = 3
blue_balls = 2
green_balls = 5

# จำนวนครั้งที่จะทำการสุ่ม
num_trials = 10

# จำนวนครั้งที่ได้ลูกบอลสีน้ำเงิน
blue_count = 0

# จำนวนครั้งที่ทำการสุ่ม
total_trials = 0

# วนลูปจนกว่าจะได้ลูกบอลสีน้ำเงิน 6 ครั้งใน 10 ครั้ง
while blue_count != 6:
    # เริ่มต้นนับจำนวนลูกบอลสีน้ำเงินในแต่ละรอบ
    blue_count = 0
    # ทำการสุ่มและนับจำนวนลูกบอลสีน้ำเงิน
    for _ in range(num_trials):
        # สุ่มลูกบอล
        ball_color = random.choice(["red", "blue", "green"])
        # ตรวจสอบสีของลูกบอล
        if ball_color == "blue":
            blue_count += 1
        # แสดงผลลัพธ์ของแต่ละรอบ
        print(f"รอบที่ {total_trials + 1}: ได้ลูกบอลสี {ball_color}")
    # เพิ่มจำนวนครั้งที่ทำการสุ่ม
    total_trials += 1

# แสดงผลลัพธ์ที่เหลือ
print("\nผลลัพธ์สุดท้าย:")
print("====================")
print(f"จำนวนครั้งที่ทำการสุ่มจนได้ลูกบอลสีน้ำเงิน 6 ครั้งใน 10 ครั้ง: ได้ {total_trials} รอบ")

# คำนวณความน่าจะเป็น total_trials คือจำนวนรอบทั้งหมดที่ทำการสุ่ม และ 10 คือจำนวนรอบที่เรากำหนดไว้ในการทดลอง
probability = total_trials / 10

# แสดงผลลัพธ์
print("ความน่าจะเป็นที่จะได้ลูกบอลสีน้ำเงิน 6 ครั้งใน 10 ครั้ง:", probability)
```
